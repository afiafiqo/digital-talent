# Deskripsi Data

```python
import pandas as pd
data_putus_sekolah_sd = pd.read_csv('data-putus-sekolah-sd.csv', delimiter = ',', encoding = "ISO-8859-1")
data_putus_sekolah_sd
```

```python
data_putus_sekolah_sd.columns
```

```python
data_putus_sekolah_sd.count()
```

Dataset diatas merupakan kumpulan data yang berisi jumlah siswa Sekolah Dasar (SD) tiap provinsi yang putus sekolah dari tahun 2008 hingga 2011. Data tersebut diambil dari situs data.go.id. Data tersebut terdiri dari 5 kolom (kode_provinsi, nama_provinsi, tahun, jumlah_putus_sekolah, dan persen_putus_sekolah). Variabel-variabel yang terdapat pada data set tersebut direpresentasikan sebagai nama kolom, dimana penjelasan dari tiap variable dijelaskan sebagai berikut :
1. kode_provinsi: Kode Provinsi 
2. nama_provinsi: Nama Provinsi 
3. tahun: Tahun 
4. jumlah_putus_sekolah: Jumlah siswa SD yang putus sekolah 
5. persen_putus_sekolah: Persentase siswa SD yang putus sekolah

# Statistik Data

```python
# Jumlah anak putus sekolah dari tahun 2008 hingga 2011 di Indonesia
print ("jumlah siswa SD yang putus sekolah dari tahun 2008 hingga 2011 di Indonesia :",data_putus_sekolah_sd.jumlah_putus_sekolah.sum(),"siswa")
```

```python
# capaian jumlah anak putus sekolah dari tahun 2008 hingga 2011 di Indonesia terendah
print ("capaian jumlah siswa SD yang putus sekolah dari tahun 2008 hingga 2011 di Indonesia terendah :",data_putus_sekolah_sd.jumlah_putus_sekolah.min(),"siswa")
```

```python
# capaian jumlah anak putus sekolah dari tahun 2008 hingga 2011 di Indonesia tertinggi
print ("capaian jumlah siswa SD yang putus sekolah dari tahun 2008 hingga 2011 di Indonesia terendah :",data_putus_sekolah_sd.jumlah_putus_sekolah.max(),"siswa")
```

```python
#jumlah siswa SD yang putus sekolah di Indonesia setiap tahun
data_putus_sekolah_sd.groupby('tahun').sum()
```

```python
#data jumlah siswa SD yang putus sekolah tiap provinsi tiap tahun
data_putus_sekolah_sd.groupby(['kode_provinsi','nama_provinsi','tahun']).sum()
```

```python
# siswa putus sekolah SD tiap provinsi dari tahun 2008 hingga 2011
data_putus_sekolah_sd.groupby(['kode_provinsi', 'nama_provinsi']).sum()
```

```python
#rata-rata siswa putus sekolah SD tiap provinsi dari tahun 2008 hingga 2011
data_putus_sekolah_sd.groupby(['kode_provinsi', 'nama_provinsi']).mean()
```

```python
#jumlah siswa SD yang putus sekolah tiap provinsi pada tahun tertentu (contoh: 2008)
data_putus_sekolah_sd[data_putus_sekolah_sd['tahun']==2008]
```

```python
#rata-rata jumlah siswa SD yang putus sekolah di Indonesia pada tahun tertentu (contoh: 2008)
data_putus_sekolah_sd[data_putus_sekolah_sd['tahun']==2008].mean()
```

```python
#jumlah siswa sd yang putus sekolah pada provinsi tertentu (contoh: Prov. Papua)
data_putus_sekolah_sd[data_putus_sekolah_sd['nama_provinsi']=="Prov. Papua"]
```

```python
#jumlah siswa SD yang putus sekolah pada provinsi tertentu pada tahun 2008 hingga 2011 (contoh: Prov. Maluku Utara)
data_putus_sekolah_sd[data_putus_sekolah_sd['nama_provinsi']=="Prov. Maluku Utara"].sum()
```

```python
#rata-rata siswa SD yang putus sekolah pada provinsi tertentu pada tahun 2008 hingga 2011 (contoh: Prov. Jawa Timur)
data_putus_sekolah_sd[data_putus_sekolah_sd['nama_provinsi']=="Prov. Jawa Timur"].mean()
```

```python
#jumlah siswa sd yang putus sekolah pada provinsi dan tahun tertentu (contoh: Prov. Bali pada tahun 2009)
data_putus_sekolah_sd.loc[(data_putus_sekolah_sd['nama_provinsi']=="Prov. Bali") & (data_putus_sekolah_sd['tahun']==2009)]
```

# Visualisasi Data

```python
# Perkembangan jumlah siswa SD yang putus sekolah di Indonesia dari tahun 2008 hingga 2011
ax = data_putus_sekolah_sd.groupby('tahun').sum()['jumlah_putus_sekolah']
ax = ax.plot(kind='bar', title ="Grafik 1 : Jumlah Berdasarkan Tahun", legend= False, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah siswa yang putus sekolah", fontsize=12)
```

```python
# Grafik jumlah siswa SD yang putus sekolah setiap provinsi setiap tahun
ax = data_putus_sekolah_sd.groupby(['nama_provinsi', 'tahun']).sum()['jumlah_putus_sekolah']
ax = ax.plot(kind='bar', title ="Grafik 2 : Jumlah Berdasarkan Provinsi dan Tahun", figsize =(30,15), legend=True, fontsize=12)
ax.set_xlabel("Provinsi, Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
# jumlah siswa SD yang putus sekolah di Indonesia pada tahun 2011
ax = data_putus_sekolah_sd[data_putus_sekolah_sd['tahun']==2011].groupby('nama_provinsi').sum()['jumlah_putus_sekolah']
ax = ax.plot(kind='bar', title ="Grafik 3 : Jumlah Berdasarkan Provinsi pada Tahun 2011", figsize=(12,12), legend= False, fontsize=12)
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("Jumlah siswa yang putus sekolah", fontsize=12)
```

```python
# jumlah siswa SD yang putus sekolah di Prov. Jawa Barat dari tahun 2008 hingga 2011
ax = data_putus_sekolah_sd[data_putus_sekolah_sd['nama_provinsi']=="Prov. Jawa Barat"].groupby('tahun').sum()['jumlah_putus_sekolah']
ax = ax.plot(kind='bar', title ="Grafik 4 : Jumlah Berdasarkan tahun pada Prov. Jawa Barat", legend= False, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah siswa yang putus sekolah", fontsize=12)
```

# Analisis

Berdasarkan pada hasil visualisasi diatas, apabila dilihat dari gambar "Grafik 1" didapatkan bahwa jumlah siswa SD yang putus sekolah di Indonesia pada tahun 2008 hingga tahun 2011 mengalami penurunan yang cukup signifikan. Hal ini dikarenakan jumlah siswa SD yang putus sekolah juga berkurang pada setiap provinsinya berdasarkan hasil visualisasi pada "Grafik 2". Namun, berdasarkan pada "Grafik 3", pada tahun 2011 Jawa Barat masih memiliki jumlah siswa SD putus sekolah yang jauh lebih tinggi dibandingkan dengan provinsi lainnya, meskipun jumlah tersebut telah mengalami penurunan dari tahun 2008 hingga 2011 yang diperoleh dari hasil visualisasi pada "Grafik 4". 
Dari uraian tersebut dapat disimpulkan bahwa persentase terbesar jumlah anak putus sekolah diperoleh dari Prov. Jawa Barat. Akan tetapi, dengan menurunnya jumlah siswa SD yang putus sekolah yang signifikan pada tahun 2011 memberikan gambaran bahwa kesadaran masyarakat Indonesia khususnya anak sekolah dasar (SD) akan pentingnya pendidikan meningkat serta berhasilnya upaya yang dilakukan pemerintah dalam mencerdaskan kehidupan bangsa melalui sekolah berjalan dengan cukup baik pada tahun 2011.
